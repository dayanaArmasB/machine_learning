```python
import os
import smtplib
import webbrowser
import speech_recognition as sr
import pyttsx3
import pywhatkit
import urllib.request
import json
import datetime
import wikipedia

# Le agregamos un nombre a nuestro asistente
name = 'viernes'

# Variable bandera que nos permite parar la ejecución del asistente
flag = 1

# Declaramos al objeto que nos permitirá convertir nuestra voz a texto
listener = sr.Recognizer()

# Declaramos al objeto que nos permite modular en voz a la respuesta en texto
engine = pyttsx3.init()

# Estas dos líneas permiten establecer una lista de voces y escoger una
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)

# Estas dos líneas permiten generar la velocidad y el volumen de la voz de respuesta
engine.setProperty('rate', 178)
engine.setProperty('volume', 1.0)

# En esta función se genera el uso del micrófono del sistema para capturar la voz, modularla a texto y guardarla en una variable
def listen():
    flag = 1
    try:
        with sr.Microphone() as source:
            print("Escuchando...")
            voice = listener.listen(source)
            rec = listener.recognize_google(voice, language='es-ES')
            rec = rec.lower()

            if name in rec:
                rec = rec.replace(name, '')
                flag = run(rec)
            else:
                flag = run(rec)
    except:
        pass
    return flag

# Es aquí donde se generan las respuestas en texto y se envían, seguidamente, a modularse en respuesta de voz dentro de la función talk()
def run(rec):
    global flag

    if 'upn ayuda' in rec:
        talk("Hola soy Viernes y estoy aquí, lista para responderte Ingeniero Americo Ramirez")
    elif 'reproduce' in rec:
        music = rec.replace('reproduce', '')
        talk('Reproduciendo ' + music)
        pywhatkit.playonyt(music)
    elif 'dime hora' in rec:
        hora = datetime.datetime.now().strftime('%I:%M %p')
        talk("Son las " + hora)
    elif 'busca' in rec:
        order = rec.replace('busca', '').replace('que es', '')
        wikipedia.set_lang("es")
        info = wikipedia.summary(order, 1)
        talk(info)
    elif 'abre excel' in rec:
        os.system("start excel")
        talk("Abriendo Excel")
    elif 'abre power point' in rec:
        os.system("start powerpnt")
        talk("Abriendo PowerPoint")
    elif 'cierra' in rec or 'close' in rec:
        flag = 0
        talk("Saliendo...")
    else:
        talk("Vuelve a intentarlo, no reconozco: " + rec)

    return flag

# Esta función se encarga de modular el texto en voz y reproducirla
def talk(text):
    engine.say(text)
    engine.runAndWait()

while flag:
    flag = listen()
```
